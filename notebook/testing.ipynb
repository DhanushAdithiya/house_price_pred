```python
le = LabelEncoder()
```

```python
cat = []
for i in df.columns:
    if not pd.api.types.is_numeric_dtype(df[i]):
        cat.append(i)
    
```

```python
cat
```

```python
for col in cat:
    df[col] = le.fit_transform(df[col])
```
